# 부산 방문지 정보 필터링 작업

```python
import pandas as pd
import numpy as np
```

```python
# Load the CSV files
file_path = './TL_csv/tn_travel_여행_F.csv'
df = pd.read_csv(file_path)

file_path_2 = './TL_csv/tn_visit_area_info_방문지정보_F.csv'
df_2 = pd.read_csv(file_path_2)

# Filter the dataframe to get TRAVEL_IDs where TRAVEL_PERSONA contains '부산'
filtered_df = df[df['TRAVEL_PERSONA'].str.contains('부산')]['TRAVEL_ID']

# Filter the second dataframe to get rows where TRAVEL_ID is in the list of filtered travel IDs
filtered_rows = df_2[df_2['TRAVEL_ID'].isin(filtered_df)]

# Remove rows where VISIT_AREA_NM or POI_NM contains '휴게소'
filtered_rows = filtered_rows[~filtered_rows['VISIT_AREA_NM'].str.contains('휴게소', na=False) & ~filtered_rows['POI_NM'].str.contains('휴게소', na=False)]

filtered_rows
```

## 필터링된 데이터프레임에서 '부산' 단어가 포함된 행 추출

```python
# Filter rows where any of the specified columns contain '부산'
filtered_rows_with_busan = filtered_rows[
    filtered_rows[['VISIT_AREA_NM', 'ROAD_NM_ADDR', 'LOTNO_ADDR', 'POI_NM']].apply(lambda x: x.str.contains('부산', na=False)).any(axis=1)
]

filtered_rows_with_busan
```

## 'VISIT_AREA_NM' 열에 정확히 '집'만 적혀있는 행 삭제

```python
# Filter out rows where VISIT_AREA_NM is exactly '집'
filtered_rows_without_home = filtered_rows_with_busan[filtered_rows_with_busan['VISIT_AREA_NM'] != '집']

filtered_rows_without_home
```

## VISIT_ORDER가 '1'이고, 해당 TRAVEL_ID가 유니크한 행 삭제
여행지가 부산이 아닌 집이 부산인 경우

```python
# Identify rows where VISIT_ORDER is '1' and TRAVEL_ID is unique
unique_travel_ids = filtered_rows_without_home['TRAVEL_ID'].value_counts()
unique_travel_ids = unique_travel_ids[unique_travel_ids == 1].index

rows_to_remove = filtered_rows_without_home[
    (filtered_rows_without_home['VISIT_ORDER'] == 1) | (filtered_rows_without_home['TRAVEL_ID'].isin(unique_travel_ids))
]

# Remove these rows from the dataframe
final_filtered_rows = filtered_rows_without_home.drop(rows_to_remove.index)

final_filtered_rows
```

## VISIT_AREA_NM, ROAD_NM_ADDR, LOTNO_ADDR, X_COORD 열이 동시에 NULL인 값 제거
위치를 알 수 없는 경우

```python
# Remove rows where ROAD_NM_ADDR, LOTNO_ADDR, and X_COORD are all NULL
final_filtered_rows_cleaned = final_filtered_rows.dropna(subset=['VISIT_AREA_NM', 'ROAD_NM_ADDR', 'LOTNO_ADDR', 'X_COORD'], how='all')

# Save the final cleaned dataframe to a CSV file with UTF-8 (BOM) encoding
output_file_path_cleaned = './부산_방문지_정보.csv'
final_filtered_rows_cleaned.to_csv(output_file_path_cleaned, index=False, encoding='utf-8-sig')

final_filtered_rows_cleaned.head()
```

## 연령별 방문 poi (경로 표시)
오른쪽 위 레이어를 통해 각 연령별로 볼 수 있음

```python
import folium
from folium import plugins
from folium.plugins import MarkerCluster

# 데이터 전처리
travel_df = cleaned_df.copy()
travel_df['TRAVEL_ID'] = travel_df['TRAVEL_ID'].str[2:]  # 앞의 두 글자 제거

# 데이터 병합
merged_df = pd.merge(travel_df, traveler_df, left_on='TRAVEL_ID', right_on='TRAVELER_ID', how='inner')
merged_df.to_csv('merge.csv', index=False, encoding='utf-8-sig')
# 4. 연령대별 경로 시각화 함수
def create_age_group_route_map(df):
    # 중심점 계산
    center_lat = df['Y_COORD'].mean()
    center_lon = df['X_COORD'].mean()
    
    # 지도 생성
    m = folium.Map(location=[center_lat, center_lon], zoom_start=7)
    
    # 색상 매핑
    age_colors = {
        '20': 'red', '30': 'blue', '40': 'green', '50': 'purple', '60': 'orange'
    }
    
    # 연령대별로 경로 그리기
    for age_group, group in df.groupby('AGE_GRP'):
        color = age_colors.get(str(age_group), 'gray')  # 정의되지 않은 연령대는 회색으로 처리
        
        # 각 연령대별 특성 그룹 생성
        fg = folium.FeatureGroup(name=f'Age Group {age_group}')
        
        for travel_id, travel_group in group.groupby('TRAVEL_ID'):
            coordinates = travel_group[['Y_COORD', 'X_COORD']].values.tolist()
            
            # 경로 그리기
            folium.PolyLine(
                coordinates, 
                color=color, 
                weight=2, 
                opacity=0.8, 
                tooltip=f'Age Group: {age_group}, Travel ID: {travel_id}'
            ).add_to(fg)
            
            # 시작점과 끝점 마커 추가
            folium.Marker(coordinates[0], popup='Start', icon=folium.Icon(color=color, icon='play')).add_to(fg)
            folium.Marker(coordinates[-1], popup='End', icon=folium.Icon(color=color, icon='stop')).add_to(fg)
        
        # 특성 그룹을 지도에 추가
        fg.add_to(m)
    
    # 레이어 컨트롤 추가
    folium.LayerControl().add_to(m)
    
    # 범례 추가
    legend_html = '''
    <div style="position: fixed; bottom: 50px; left: 50px; width: 120px; height: 180px; 
    border:2px solid grey; z-index:9999; font-size:14px; background-color:white">
    <p><strong>Age Groups</strong></p>
    '''
    for age, color in age_colors.items():
        legend_html += f'<p><span style="color:{color};">■</span> {age}s</p>'
    legend_html += '</div>'
    m.get_root().html.add_child(folium.Element(legend_html))
    
    return m

# 5. 지도 생성
age_group_route_map = create_age_group_route_map(merged_df)

# 6. 지도 저장
age_group_route_map.save('age_group_travel_routes.html')
```

## 연령 별 방문 POI (경로 제외, POI 점만)

## 연령별 방문 장소의 분산
나이가 많을 수록 자차를 많이 보유하고 있으니 분산이 넓을 것이라 생각했지만 비례하진 않았음

```python
def calculate_dispersion(group):
    lat_var = np.var(group['Y_COORD'])
    lon_var = np.var(group['X_COORD'])
    return lat_var + lon_var

# 각 연령대별로 방문 장소의 분산 계산
age_group_dispersion = merged_df.groupby('AGE_GRP').apply(calculate_dispersion).sort_index()

# 결과 출력
print("연령대별 방문 장소의 분산:")
for age_group, dispersion in age_group_dispersion.items():
    print(f"연령대 {age_group}: {dispersion:.6f}")

# 시각화를 위한 막대 그래프 생성
plt.figure(figsize=(12, 6))
bars = plt.bar(age_group_dispersion.index.astype(str), age_group_dispersion.values)

plt.title('Dispersion of Visited Locations by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Dispersion (Latitude + Longitude Variance)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 각 막대 위에 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.6f}',
             ha='center', va='bottom', fontsize=10, rotation=0)

plt.tight_layout()
plt.savefig('age_group_location_dispersion.png')
plt.close()
```

## 클러스터링

```python
from sklearn.cluster import KMeans

def create_cluster_map(group, age_group, n_clusters=10):
    # 클러스터링
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(group[['X_COORD', 'Y_COORD']])
    group['Cluster'] = kmeans.labels_

    # 중심점 계산
    center_lat = group['Y_COORD'].mean()
    center_lon = group['X_COORD'].mean()
    
    # 지도 생성
    m = folium.Map(location=[center_lat, center_lon], zoom_start=7)

    # 클러스터별 색상
    colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'yellow', 'darkblue', 'darkgreen']

    # 클러스터별로 마커 추가
    for i in range(n_clusters):
        cluster_points = group[group['Cluster'] == i]
        for idx, row in cluster_points.iterrows():
            folium.CircleMarker(
                location=[row['Y_COORD'], row['X_COORD']],
                radius=3,
                popup=row['VISIT_AREA_NM'],
                color=colors[i % len(colors)],
                fill=True,
                fillColor=colors[i % len(colors)],
                fillOpacity=0.7
            ).add_to(m)

    # 클러스터 중심점 추가
    cluster_centers = kmeans.cluster_centers_
    for i, center in enumerate(cluster_centers):
        folium.Marker(
            location=[center[1], center[0]],
            icon=folium.Icon(color=colors[i % len(colors)], icon='info-sign'),
            popup=f'Cluster {i+1} Center'
        ).add_to(m)

    # 범례 추가
    legend_html = '''
    <div style="position: fixed; bottom: 50px; left: 50px; width: 150px; 
    border:2px solid grey; z-index:9999; font-size:14px; background-color:white;">
    <p><strong>Clusters</strong></p>
    '''
    for i in range(n_clusters):
        legend_html += f'<p><span style="color:{colors[i % len(colors)]};">●</span> Cluster {i+1}</p>'
    legend_html += '</div>'
    m.get_root().html.add_child(folium.Element(legend_html))

    return m

# 각 연령대별로 클러스터 지도 생성
for age_group, group in merged_df.groupby('AGE_GRP'):
    m = create_cluster_map(group, age_group)
    m.save(f'age_group_{age_group}_clusters.html')

print("각 연령대별 클러스터 지도가 생성되었습니다.")
```

## 성별 방문 장소

```python
def create_gender_visit_map(df):
    # 중심점 계산
    center_lat = df['Y_COORD'].mean()
    center_lon = df['X_COORD'].mean()
    
    # 지도 생성
    m = folium.Map(location=[center_lat, center_lon], zoom_start=7)
    
    # 색상 매핑
    gender_colors = {
        '남': 'blue', '여': 'red'
    }
    
    # 성별별 방문지 수 계산
    gender_counts = df['GENDER'].value_counts()
    
    # 성별별 방문지 수 출력
    print(f"남자 방문지 수: {gender_counts.get('남', 0)}")
    print(f"여자 방문지 수: {gender_counts.get('여', 0)}")
    
    # 성별별 마커 그룹 생성
    for gender in df['GENDER'].unique():
        gender_data = df[df['GENDER'] == gender]
        color = gender_colors.get(gender, 'gray')
        
        # 마커 그룹 생성
        fg = folium.FeatureGroup(name=f'Gender {gender}')
        
        for _, row in gender_data.iterrows():
            folium.CircleMarker(
                location=[row['Y_COORD'], row['X_COORD']],
                radius=5,
                popup=f"Gender: {gender}, Location: {row['VISIT_AREA_NM']}",
                color=color,
                fill=True,
                fillColor=color
            ).add_to(fg)
        
        fg.add_to(m)
    
    # 레이어 컨트롤 추가
    folium.LayerControl().add_to(m)
    
    # 범례 추가
    legend_html = '''
    <div style="position: fixed; bottom: 50px; left: 50px; width: 120px; height: 90px; 
    border:2px solid grey; z-index:9999; font-size:14px; background-color:white">
    <p><strong>Gender</strong></p>
    <p><span style="color:blue;">●</span> 남</p>
    <p><span style="color:red;">●</span> 여</p>
    </div>
    '''
    m.get_root().html.add_child(folium.Element(legend_html))
    
    return m

# 지도 생성
gender_visit_map = create_gender_visit_map(merged_df)

# 지도 저장
gender_visit_map.save('gender_travel_locations.html')
```

## 월별 방문 장소
4~9월 데이터가 존재  
4월은 데이터가 3일밖에 존재하지 않음

```python
def create_monthly_visit_map(df):
    # VISIT_START_YMD 열에서 월을 추출
    df['VISIT_MONTH'] = pd.to_datetime(df['VISIT_START_YMD']).dt.month
    
    # 중심점 계산
    center_lat = df['Y_COORD'].mean()
    center_lon = df['X_COORD'].mean()
    
    # 지도 생성
    m = folium.Map(location=[center_lat, center_lon], zoom_start=7)
    
    # 색상 매핑
    month_colors = {
        4: 'red', 5: 'blue', 6: 'green', 7: 'purple', 8: 'orange', 9: 'darkred'
    }
    
    # 월별 마커 그룹 생성
    for month in df['VISIT_MONTH'].unique():
        month_data = df[df['VISIT_MONTH'] == month]
        color = month_colors.get(month, 'gray')
        
        # 마커 그룹 생성
        fg = folium.FeatureGroup(name=f'Month {month}')
        
        for _, row in month_data.iterrows():
            folium.CircleMarker(
                location=[row['Y_COORD'], row['X_COORD']],
                radius=5,
                popup=f"Month: {month}, Location: {row['VISIT_AREA_NM']}",
                color=color,
                fill=True,
                fillColor=color
            ).add_to(fg)
        
        fg.add_to(m)
    
    # 레이어 컨트롤 추가
    folium.LayerControl().add_to(m)
    
    # 범례 추가
    legend_html = '''
    <div style="position: fixed; bottom: 50px; left: 50px; width: 140px; height: 280px; 
    border:2px solid grey; z-index:9999; font-size:14px; background-color:white">
    <p><strong>Months</strong></p>
    '''
    for month, color in month_colors.items():
        legend_html += f'<p><span style="color:{color};">●</span> {month}월</p>'
    legend_html += '</div>'
    m.get_root().html.add_child(folium.Element(legend_html))
    
    return m

# 지도 생성
monthly_visit_map = create_monthly_visit_map(merged_df)

# 지도 저장
monthly_visit_map.save('monthly_travel_locations.html')
```

## 시청 기준 동서 나누기

```python
def create_cityhall_east_west_map(df):
    # 기준점 설정 (시청)
    cityhall_lon = 129.0768832
    
    # 중심점 계산
    center_lat = df['Y_COORD'].mean()
    center_lon = df['X_COORD'].mean()
    
    # 지도 생성
    m = folium.Map(location=[center_lat, center_lon], zoom_start=11)
    
    # 동서로 나누기
    east_data = df[df['X_COORD'] > cityhall_lon]
    west_data = df[df['X_COORD'] <= cityhall_lon]
    
    # 방문 지점 개수 출력
    print(f'시청 기준 동쪽 방문 지점 개수: {len(east_data)}')
    print(f'시청 기준 서쪽 방문 지점 개수: {len(west_data)}')
    
    # 동쪽 마커 그룹 생성
    east_fg = folium.FeatureGroup(name='East of City Hall')
    for _, row in east_data.iterrows():
        folium.CircleMarker(
            location=[row['Y_COORD'], row['X_COORD']],
            radius=5,
            popup=f"Location: {row['VISIT_AREA_NM']}",
            color='blue',
            fill=True,
            fillColor='blue'
        ).add_to(east_fg)
    
    # 서쪽 마커 그룹 생성
    west_fg = folium.FeatureGroup(name='West of City Hall')
    for _, row in west_data.iterrows():
        folium.CircleMarker(
            location=[row['Y_COORD'], row['X_COORD']],
            radius=5,
            popup=f"Location: {row['VISIT_AREA_NM']}",
            color='red',
            fill=True,
            fillColor='red'
        ).add_to(west_fg)
    
    # 특성 그룹을 지도에 추가
    east_fg.add_to(m)
    west_fg.add_to(m)
    
    # 레이어 컨트롤 추가
    folium.LayerControl().add_to(m)
    
    # 범례 추가
    legend_html = '''
    <div style="position: fixed; bottom: 50px; left: 50px; width: 150px; height: 90px; 
    border:2px solid grey; z-index:9999; font-size:14px; background-color:white">
    <p><strong>City Hall</strong></p>
    <p><span style="color:blue;">●</span> East</p>
    <p><span style="color:red;">●</span> West</p>
    </div>
    '''
    m.get_root().html.add_child(folium.Element(legend_html))
    
    return m

# 지도 생성
cityhall_east_west_map = create_cityhall_east_west_map(merged_df)

# 지도 저장
cityhall_east_west_map.save('cityhall_east_west_travel_locations.html')
```

## 부산항 기준으로 동서 나누기

```python
def create_port_east_west_map(df):
    # 기준점 설정 (부산항)
    port_lon = 129.0490206
    
    # 중심점 계산
    center_lat = df['Y_COORD'].mean()
    center_lon = df['X_COORD'].mean()
    
    # 지도 생성
    m = folium.Map(location=[center_lat, center_lon], zoom_start=11)
    
    # 동서로 나누기
    east_data = df[df['X_COORD'] > port_lon]
    west_data = df[df['X_COORD'] <= port_lon]
    
    # 방문 지점 개수 출력
    print(f'부산항 기준 동쪽 방문 지점 개수: {len(east_data)}')
    print(f'부산항 기준 서쪽 방문 지점 개수: {len(west_data)}')
    
    # 동쪽 마커 그룹 생성
    east_fg = folium.FeatureGroup(name='East of Port')
    for _, row in east_data.iterrows():
        folium.CircleMarker(
            location=[row['Y_COORD'], row['X_COORD']],
            radius=5,
            popup=f"Location: {row['VISIT_AREA_NM']}",
            color='blue',
            fill=True,
            fillColor='blue'
        ).add_to(east_fg)
    
    # 서쪽 마커 그룹 생성
    west_fg = folium.FeatureGroup(name='West of Port')
    for _, row in west_data.iterrows():
        folium.CircleMarker(
            location=[row['Y_COORD'], row['X_COORD']],
            radius=5,
            popup=f"Location: {row['VISIT_AREA_NM']}",
            color='red',
            fill=True,
            fillColor='red'
        ).add_to(west_fg)
    
    # 특성 그룹을 지도에 추가
    east_fg.add_to(m)
    west_fg.add_to(m)
    
    # 레이어 컨트롤 추가
    folium.LayerControl().add_to(m)
    
    # 범례 추가
    legend_html = '''
    <div style="position: fixed; bottom: 50px; left: 50px; width: 150px; height: 90px; 
    border:2px solid grey; z-index:9999; font-size:14px; background-color:white">
    <p><strong>Port</strong></p>
    <p><span style="color:blue;">●</span> East</p>
    <p><span style="color:red;">●</span> West</p>
    </div>
    '''
    m.get_root().html.add_child(folium.Element(legend_html))
    
    return m

# 지도 생성
port_east_west_map = create_port_east_west_map(merged_df)

# 지도 저장
port_east_west_map.save('port_east_west_travel_locations.html')
```

```python
# 방문 여부를 나타내는 새로운 열 추가
merged_df['VISITED'] = 1

# 피벗 테이블 생성
visit_matrix = merged_df.pivot_table(index='POI_ID', columns='TRAVEL_ID', values='VISITED', fill_value=0)
visit_matrix
```

### 각 POI별 유사도 top k개 출력

```python
from sklearn.metrics.pairwise import cosine_similarity
# 코사인 유사도 계산
cosine_sim = cosine_similarity(visit_matrix)

# 결과를 데이터프레임으로 변환
cosine_sim_df = pd.DataFrame(cosine_sim, index=visit_matrix.index, columns=visit_matrix.index)
cosine_sim_df
```

```python
import numpy as np

# 코사인 유사도 계산
cosine_sim = cosine_similarity(visit_matrix)

# 결과를 데이터프레임으로 변환
cosine_sim_df = pd.DataFrame(cosine_sim, index=visit_matrix.index, columns=visit_matrix.index)

# 0을 제외한 가장 작은 값 찾기
non_zero_min = cosine_sim_df[cosine_sim_df != 0].min().min()

# 1을 제외한 가장 큰 값 찾기(1에 근사한 값이 있어서 아래와 같이 처리)
non_one_max = cosine_sim_df[cosine_sim_df < 0.999].max().max()

print("최소값(0제외):",non_zero_min, "최대값(1제외):", non_one_max)
```

```python
# POI별로 top k 유사한 POI 출력 (자기 자신 제외)
top_k = 5
top_k_similar_pois = {}

for poi in cosine_sim_df.index:
    similar_pois = cosine_sim_df[poi].drop(poi).sort_values(ascending=False).head(top_k).index
    top_k_similar_pois[poi] = similar_pois.tolist()

# 결과 출력
top_k_similar_pois
```

### poi 유사도 시각화
각 poi를 클릭하면 유사한 k개의 정보가 나옴

```python
def get_poi_info(poi_id):
    poi_rows = merged_df[merged_df['POI_ID'] == poi_id]
    if not poi_rows.empty:
        row = poi_rows.iloc[0]
        area_name = row['VISIT_AREA_NM']
        road_name = row['ROAD_NM_ADDR'] if pd.notna(row['ROAD_NM_ADDR']) else row['LOTNO_ADDR'] if pd.notna(row['LOTNO_ADDR']) else ''
        return area_name, road_name
    return None, None

# 지도 중심점 계산
center_lat = merged_df['Y_COORD'].mean()
center_lon = merged_df['X_COORD'].mean()

# 지도 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# POI 마커 추가
for poi in merged_df['POI_ID'].unique():
    poi_data = merged_df[merged_df['POI_ID'] == poi]
    if not poi_data.empty:
        poi_data = poi_data.iloc[0]
        lat = poi_data['Y_COORD']
        lon = poi_data['X_COORD']
        area_name, road_name = get_poi_info(poi)
        
        # 팝업 내용 생성
        if area_name and road_name:
            popup_content = f'<b>POI {poi}</b><br><b>Area:</b> {area_name}<br><b>Address:</b> {road_name}<br><br>'
            popup_content += '<b>Top K Similar POIs:</b><br>'
            
            for similar_poi in top_k_similar_pois[poi]:
                sim_area_name, sim_road_name = get_poi_info(similar_poi)
                if sim_area_name and sim_road_name:
                    popup_content += f'{similar_poi}: {sim_area_name}, {sim_road_name}<br>'
            
            folium.Marker(
                location=[lat, lon],
                popup=folium.Popup(popup_content, max_width=300)
            ).add_to(m)

# 지도 저장
m.save('poi_similarity_map1.html')
```

```python
import colorlover as cl

# 중심 좌표 계산
center_lat = merged_df['Y_COORD'].mean()
center_lon = merged_df['X_COORD'].mean()

# 지도 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# 상위 n개 POI 선택
top_n_pois = merged_df['POI_ID'].unique()[:5]

# 모든 POI 정보 딕셔너리 생성
all_poi_info = merged_df.groupby('POI_ID').first()[['POI_NM', 'X_COORD', 'Y_COORD']].to_dict('index')

# 각 POI에 대한 마커 추가
for poi_id, info in all_poi_info.items():
    color = 'red' if poi_id in top_n_pois else 'blue'
    folium.CircleMarker(
        location=[info['Y_COORD'], info['X_COORD']],
        radius=5,
        popup=info['POI_NM'],
        color=color,
        fill=True,
        fillColor=color,
        fillOpacity=0.7,
        tooltip=info['POI_NM'],
    ).add_to(m)

# n가지 색상 생성
colors = cl.scales['5']['qual']['Paired']

# 상위 n개 POI에 대해 유사한 POI와 엣지 연결
for i, poi_id in enumerate(top_n_pois):
    poi = all_poi_info[poi_id]
    similar_pois = top_k_similar_pois[poi_id]
    color = colors[i]
    
    for similar_poi_id in similar_pois:
        similar_poi = all_poi_info[similar_poi_id]
        
        folium.PolyLine(
            locations=[
                [poi['Y_COORD'], poi['X_COORD']],
                [similar_poi['Y_COORD'], similar_poi['X_COORD']]
            ],
            color=color,
            weight=2,
            opacity=0.8
        ).add_to(m)

# 지도 저장
m.save('poi_similarity_map2.html')
```
